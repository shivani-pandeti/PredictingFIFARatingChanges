## WEBSCRAPER 1

This is the working Web Scraper - 1 person (Kylian Mbappe)

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

def get_player_ratings(url):
    # Set up Selenium WebDriver
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)
    driver.get(url)
    
    # Get page source and parse with BeautifulSoup
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    player_ratings = {}
    
    # Find the table with the ratings
    table = soup.find('table', class_='table table-sm table-striped table-hover mb-0')
    if table:
        rows = table.find('tbody').find_all('tr')
        for row in rows[1:]:  # Skip the header row
            columns = row.find_all('td')
            game = columns[0].get_text(strip=True)
            rating = columns[1].get_text(strip=True)
            player_ratings[game] = rating
    else:
        print("Table not found.")
    
    driver.quit()
    return player_ratings

url = "https://www.fifaratings.com/nathan-ake"
ratings = get_player_ratings(url)
print(ratings)
```

## WEB SCRAPER 2

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

def get_player_ratings(url):
    # Set up Selenium WebDriver
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)
    driver.get(url)
    
    # Get page source and parse with BeautifulSoup
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    player_ratings = {}
    
    # Find the table with the ratings
    table = soup.find('table', class_='table table-sm table-striped table-hover mb-0')
    if table:
        rows = table.find('tbody').find_all('tr')
        for row in rows[1:]:  # Skip the header row
            columns = row.find_all('td')
            game = columns[0].get_text(strip=True)
            rating = columns[1].get_text(strip=True)
            player_ratings[game] = rating
    else:
        print("Table not found.")
    
    driver.quit()
    return player_ratings

df_all_ratings = pd.DataFrame()

for name in final_names:
    url = f"https://www.fifaratings.com/{name}"
    ratings = get_player_ratings(url)
    # Convert the ratings dictionary to a DataFrame and add the player's name
    df_ratings = pd.DataFrame(ratings, index=[0])
    df_ratings['Name'] = name
    # Append to the main DataFrame
    df_all_ratings = pd.concat([df_all_ratings, df_ratings], ignore_index=True)
    time.sleep(30)
```

## Remove accents from names 

```python
from unidecode import unidecode

# Sample string with accented characters
text_with_accents = df_all_ratings['Name'].tolist()

# Replace accented characters with non-accented equivalents
text_without_accents = [unidecode(text) for text in text_with_accents]

print(text_without_accents)
```

```python
df_all_ratings = df_all_ratings.assign(name_without_accents=text_without_accents)
df_all_ratings.info()
```

```python
names_with_null_fc25 = df_all_ratings[df_all_ratings['FC 25'].isnull()]['name_without_accents'].tolist()
```
